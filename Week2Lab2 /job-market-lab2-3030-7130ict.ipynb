```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load job market
jobmarket = pd.read_csv('/Users/amalnormanhakim/Desktop/3030ICT - Data Analytics/Lab2/job-market(3).csv')
jobmarket.head()
```

```python
%matplotlib inline
#Visualise Job by Location
fig, ax = plt.subplots(figsize=(10, 20))
ax = sns.countplot(data=jobmarket, y="Location", order = jobmarket['Location'].value_counts().index)
ax.set_title("Number of Jobs by Location")
ax.set_xlabel("Number of Jobs")
ax.set_ylabel("Location");
```

```python
import matplotlib
#create dataframe for salary ranges
salary = jobmarket.groupby("LowestSalary").count()
salary.head()
```

```python
#salary ranges plot
fig, ax = plt.subplots(figsize=(6, 6))
labels = salary.index
ax.pie(salary['Id'], labels = labels,  autopct='%.1f%%')
ax.set_title('Job Posts by Salary Range')
```
